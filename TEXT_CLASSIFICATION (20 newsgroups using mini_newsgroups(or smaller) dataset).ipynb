```python
#function to remove irrelevant characters like "," ,":" e.t.c from a string
 
def remove_irrelevant_characters(str):
    new_str = ""
    for c in str:
        if(c == '.' or (ord(c) >= 65 and ord(c) <= 90) or (ord(c) >= 97 and ord(c) <= 122)):
            new_str += c 
    if(len(str) == 1):
        return ""
    return new_str.lower()        


```

```python

#"word" function is to check that string has atleast one alphabet
def word(str):
    for c in str:
        if(ord(c) >= 65 and ord(c) <= 90) or (ord(c) >= 97 and ord(c) <= 122):
            return True
    return False

#"add_file_to_dict" function take input the file_path and add all the relevant words of that file to word_dictionary 
def add_file_to_dict(file_path):
    f = open(file_path,'r')
    message = f.read()
    message2 = message.split()
    for line in message2:
            if(len(line) == 0):
                continue
            str = remove_irrelevant_characters(line) 
            if(str in word_dict and str not in stopwords and len(str) > 0 and word(str)):
                word_dict[str] += 1
            elif(str not in stopwords and len(str) > 0 and word(str)):
                word_dict[str] = 1
    f.close()
```

```python
#code for fit and predict function


#with laplace corrrection
#and converting ab to log(a) + log(b)

#fit function to fit the training data
def fit2(x_train,y_train):
    result = {} 
    class_values = set(y_train)
    result["total_data"] = len(y_train)
    for current_class in class_values:
        result[current_class] = {} 
        current_class_rows = (y_train == current_class)
        x_train_current = x_train[current_class_rows] 
        y_train_current = y_train[current_class_rows] 
        num_features = x_train.shape[1]
        result[current_class]["total_count"] = len(x_train_current)
        for j in range(1,num_features+1):
            result[current_class][j-1] = x_train_current[:,j-1].sum() 
    return result 


#probability function predicts the probability that a test doc. belong to given class
#here we only calculate probabilities for those words which lies atleast once in testing data(i.e., x[j-1] > 0)
def probability2(dictionary, x, current_class):
    output = 1
    class_probab = np.log(dictionary[current_class]["total_count"]) - np.log(dictionary["total_data"]) # p(y = ai) = count(y = ai)/total_data
    output = class_probab
    num_features = len(features)
    for j in range(1,num_features+1):
        if(x[j-1] > 0):
            count_current_class_with_value_xj = dictionary[current_class][j-1] + 1 # with laplace correction numerator i.e., count(jth feature = xj && y = current_class)
            count_current_class = dictionary[current_class]["total_count"] + num_features#denominator i.e., count(y = current_class)
            current_xj_probability = np.log(count_current_class_with_value_xj) - np.log(count_current_class)
            output = output + current_xj_probability
    return output    
                                                                             
#here we run over all 20 available classes and select that class for which probability is maximum                                                                         
def predictsinglepoint2(dictionary, x):
    classes = dictionary.keys()
    best_p = -1000
    best_class = -1
    first_run = True
    for current_class in classes:
        if(current_class == "total_data"):
            continue                                                                 
        p_current_class = probability2(dictionary,x,current_class) #probab that x data will belong to current_class
        if(first_run or p_current_class > best_p):
            best_p = p_current_class
            best_class = current_class
        first_run = False;    
    return best_class    
 
#predict function we run over all the document in x_test and find the most appropriate class for each document     
def predict2(dictionary,x_test):
    y_pred = []
    for x in x_test:
        x_class = predictsinglepoint2(dictionary, x)
        y_pred.append(x_class)
    return y_pred    
```
